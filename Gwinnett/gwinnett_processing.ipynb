```python
import pandas as pd, os, re
```

```python
# using os to grab the filesets needed
appeals_data_fp = '/Users/bartholemew/urban_research_group/gwinnett_organized/appeals_data'
ownership_data_fp = '/Users/bartholemew/urban_research_group/gwinnett_organized/sales_data'

appeals_docs = os.listdir(appeals_data_fp)
ownership_docs = os.listdir(ownership_data_fp)

# fixing ownership doc names

# for name in ownership_docs:
#     year = re.search(r'\d{4}',name)
#     old_name = f"{ownership_data_fp}/{name}"
#     new_name = f"{ownership_data_fp}/Gwinnett {year.group().strip()} Ownership Data.xlsx"
#     os.rename(old_name, new_name)

appeals_docs = os.listdir(appeals_data_fp)
ownership_docs = os.listdir(ownership_data_fp)
        
```

```python
# parsing through data to find columns for var mapping
# creating a df from each excel file
appeals_df_list = []
for document in appeals_docs:
        print(f"{appeals_data_fp}/{document}")
        year = re.search(r'\d{4}',document)
        year = int(year.group().strip())
        df = pd.read_excel(f"{appeals_data_fp}/{document}")

        # add tax_yr column
        df['tax_yr'] = year  # Add the year column

        headers = df.columns.tolist()
        print({year:headers})
        appeals_df_list.append((year,df))

ownership_df_list = []
for document in ownership_docs:
        year = re.search(r'\d{4}',document) # grab year
        year = int(year.group().strip())
        df = pd.read_excel(f"{ownership_data_fp}/{document}")

        # need to add column with tax_yr
        df['tax_yr'] = year

        headers = df.columns.tolist() # headers in this set of files differ after a certain year, check which year it is
        print({year:headers})
        ownership_df_list.append((year,df))
```

```python
# print([item[0] for item in ownership_df_list]) # missing 2022, RANGE IS 2011-2021
# print([item[0] for item in appeals_df_list]) # has 2011-2022

ownership_df_list.sort(key=lambda x: x[0])
appeals_df_list.sort(key=lambda x: x[0])

for year, df in ownership_df_list:
    print(f"\nOwnership Data Columns for {year}:\n")
    print(len(list(df.columns)),"\n")
    print(df.columns)

for year, df in appeals_df_list:
    print(f"\nAppeals Columns for {year}:\n")
    print(len(list(df.columns)),"\n")
    print(df.columns)
```

```python
# drop columns that aren't needed

ownership_drop_cols = ['LRSNum','DOC1REF', 'DOC2REF','DOC3REF', 'DISTNUM','DISTNUM_DESC','PROPCLAS_DESC','LEGAL1','LEGALAC','stories', 'yrbuilt', 'FinSize','stories', 'yrbuilt', 'EXEMPT1','Neighborhood #','Public_NeiNum','PROPERTYState','LocState','State']

for year,df in ownership_df_list:
    df.drop(ownership_drop_cols, axis=1, inplace=True, errors='ignore')

for year, df in ownership_df_list:
    df.rename(columns={'PROPERTY ADDR':'PROPERTYSTREET',
    'PROPERTY CITY':'PROPERTYCity',
    'PROPERTY ZIP':'PROPERTYZip',
    'OWNER1':'OWNERNAME1',
    'OWNER2':'OWNERNAME2',
    'MAILADDR':'OWNERADDRESS1',
    'MAILCITY':'OWNERCITY',
    'MAILSTAT':'OWNERSTATE',
    'MAILZIP':'OWNERZIP',
    'LEGALAC':'LEGAL1',
    'PROPERTY DESCRIPTION':'PCDESC',
    'ZONING TYPE':'ZONEDESC',
    'GRANTOR1':'GRANTORNAME1', 
    'GRANTOR2':'GRANTORNAME2', 
    'GRANTOR3':'GRANTORNAME3',
    'Property Address':'PROPERTYSTREET',
    'Property City':'PROPERTYCity',
    'PROP_ZIP':'PROPERTYZip',
    'LOCADDR':'PROPERTYSTREET',
    'LocCity':'PROPERTYCity', 
    'LocZip':'PROPERTYZip',
    }, inplace=True)


for year,df in appeals_df_list:
    df.drop('LRSN', axis=1, inplace=True, errors='ignore')
```

```python
# create final merged dfs
appeals_dfs = [i[1] for i in appeals_df_list]
appeals_df = pd.concat(appeals_dfs, ignore_index=True)

ownership_dfs = [i[1] for i in ownership_df_list]
ownership_df = pd.concat(ownership_dfs, ignore_index=True)
```
